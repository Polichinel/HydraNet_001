```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
with open('/home/number_one/Documents/scripts/conflictNet/data/raw/viewser_monthly_vol.pkl', 'rb') as f:
    vol = pickle.load(f)
```
